# LAlatex from Python

This notebook uses `juliacall` to call Julia's `LAlatex` from a Python kernel and render the LaTeX output.

```python
from juliacall import Main as jl
from IPython.display import Latex, display

jl.seval("using LAlatex")

latex1 = jl.LAlatex.L_show("A = ", [[1, 2], [3, 4]])
latex2 = jl.LAlatex.L_show("x = ", 3, "; ", "x^2 = ", 9)

display(Latex(latex1))
display(Latex(latex2))
```
