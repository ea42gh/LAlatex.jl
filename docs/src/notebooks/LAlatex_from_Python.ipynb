# LAlatex from Python

This notebook uses `juliacall` to call Julia's `LAlatex` from a Python kernel and render the LaTeX output.

The Python helpers `l_show(...)` and `L(...)` shown below come from an external startup file named `10-julia-magic.py` in the `elementary-linear-algebra` notebook environment. That file is not part of `LAlatex`.

```python
%%julia
using LAlatex
A = [1 2 3; 4 5 6]
display(l_show(L"A = ", A, L",\quad A^T A = ", A'A))
```

```python
from juliacall import Main as jl
jl.seval("using LAlatex")

A = [[1, 2, 4], [3, 4, 1]]
l_show("A = ", A)
l_show("x = ", 3, L(r";\quad "), "x^2 = ", 9)
```
