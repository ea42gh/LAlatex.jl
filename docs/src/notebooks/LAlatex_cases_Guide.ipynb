# LAlatex cases guide

This notebook documents `cases(...)`, the structured helper for piecewise definitions and case-by-case linear algebra displays.

Use `cases` when you want LAlatex to build a LaTeX `cases` environment from Julia values. Each value and condition is rendered through the normal `L_show` policy, so vectors, matrices, symbolic expressions, strings, and `LaTeXString`s behave the same way they do elsewhere in LAlatex.

````julia
using LAlatex
using LaTeXStrings

set_backend!(:symbolics)
@syms x y t
````

## Entry forms

`cases` accepts two equivalent entry forms:

- `value => condition`
- `(value, condition)`

The pair form is compact for scalar cases. The tuple form is useful when the value or condition is already produced by another expression.

````julia
display(l_show(
    "f(x) = ",
    cases(
        x^2 => L"x \ge 0",
        (-x, L"x < 0"),
    ),
))
````

## Implicit comma and alignment

`cases` inserts the case separator and alignment marker for you. A row is rendered as:

```text
value, & condition \\
```

Do not put `&` in your input. Pass the value and the condition as separate cells.

## Text versus math cells

Plain Julia strings are rendered as text with `\text{...}`. Use `L"..."` or `LaTeXString` when the condition is already LaTeX math.

````julia
display(l_show(
    cases(
        0 => "otherwise",
        1 => L"v \in \mathcal{N}(A)",
    ),
))
````

## Vector and matrix values

The value side is a normal LAlatex display cell. This makes `cases` useful for piecewise linear maps, projections, and basis-dependent definitions.

````julia
display(l_show(
    L"T(v) = ",
    cases(
        [x, 0] => L"v \in \operatorname{span}\{e_1\}",
        [0, y] => L"v \in \operatorname{span}\{e_2\}",
        ([x, y], "otherwise"),
    ),
    arraystyle=:bmatrix,
))
````

## Propagated display options

`cases` does not have separate formatting keywords of its own. It propagates the relevant `L_show` options into every value and condition cell:

- `arraystyle`: vector and matrix delimiters inside case values
- `number_formatter`: scalar number formatting inside case values
- `per_element_style`: matrix entry styling inside case values
- `factor_out`: denominator factoring for array values
- `symopts`: display-time symbolic transforms
- `color`: color wrapper for each rendered cell

````julia
display(l_show(
    cases(
        (x + y)^2 => L"x > 0",
        [1//2, 1//3] => "rational vector",
    );
    symopts=(expand=true,),
    arraystyle=:bmatrix,
    number_formatter=n -> n isa Integer ? LaTeXString("\\mathbf{$n}") : n,
))
````

## Invalid entries

Every entry must be either a pair or a two-tuple. A bare value is rejected because LAlatex cannot infer the missing condition.

````julia
# This throws ArgumentError:
# L_show(cases(x))
````

## Summary

Use `cases(value => condition, ...)` or `cases((value, condition), ...)`. Let `cases` insert the comma and alignment marker. Use plain strings for text conditions and `LaTeXString`s for raw math. Use normal `L_show` keywords to control the rendering of values and conditions.
