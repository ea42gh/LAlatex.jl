# L_show and l_show guide

This notebook documents LAlatex's LaTeX display helpers, focused on `L_show` and `l_show`.
- `L_show` returns a raw LaTeX string.
- `l_show` returns a `LaTeXString` for notebook display.

```julia
using LAlatex
using LinearAlgebra, SparseArrays, BlockArrays
using LaTeXStrings, Symbolics
```

## Scalars

```julia
L_show(3)
display(l_show(3))
```

## Strings and LaTeXStrings

```julia
println("latex: ", L_show("x = ", L"\alpha"))
l_show("x = ", L"\alpha")
```

## Tuples and sets

```julia
g = set(1, 2, 3; separator=L", ")
l_show(g)
```

## Vectors and linear combinations

```julia
s = [2, -1, 3]
X = [[1, 0], [0, 1], [1, 1]]
l_show(lc(s, X))
```

## Standard matrices

```julia
A = [1 2; 3 4]
L_show(A; arraystyle=:bmatrix)
l_show(A; arraystyle=:pmatrix)
```

## Complex and Symbolics matrices

```julia
C = [1+2im 3-4im]
l_show(C)
```

```julia
using LAlatex:@syms  # allows backend switching
@syms x y
S = [x y; y x]
l_show(S)
```

## Special matrices

```julia
display(l_show(Diagonal([1, 2, 3])))
display(l_show(sprand(3, 3, 0.4)))
```

## BlockArrays

```julia
B = BlockArray([1 2; 3 4], [1, 1], [1, 1])
display(l_show(B; arraystyle=:barray))
```

## Per-element formatting

```julia
pivots = [1, 3]
display(l_show(A; per_element_style=(x,i,j,fx) -> echelon_pivot_formatter(x, i, j, fx; pivots=pivots)))
```

## Combining different types

```julia
b = [5, 6]
display(l_show("A = ", A, " , b = ", b))
```

## Inline vs block LaTeX

```julia
L_show(A; inline=true)
L_show(A; inline=false)
```

## Raw LaTeX output

```julia
raw = L_show(A)
raw
```

## Python display helper (optional)

```julia
try
    py_show("A = ", A)
catch err
    err
end
```
