# LAlatex basics

This notebook shows backend switching, Symbolics assumptions, SymPy helpers, and HTML helpers.

```julia
using LAlatex
using LaTeXStrings
```

## Backend switching

```julia
LAlatex.@syms x y
set_backend!(:sympy)
LAlatex.@syms a :real => true :positive => true
set_backend!(:symbolics)
LAlatex.@syms b
@show x,y,a,b
```

## Symbolics assumptions (metadata)

```julia
LAlatex.@syms t :real => true
assumptions(t)
```

## SymPy helpers

```julia
@syms_sympy p :real => true :positive => true
q = syms_sympy(:q)
```

## LaTeX helpers

```julia
to_latex(3//4)
```

### to_latex with Symbolics and SymPy

```julia
set_backend!(:symbolics)
LAlatex.@syms s
println( to_latex(s) )

set_backend!(:sympy)
sp = syms(:s; backend=:sympy, real=true)
to_latex(sp)
```
