# LAlatex basics

This notebook shows backend switching, Symbolics assumptions, SymPy helpers, and HTML helpers.

What you will learn

- Switch between Symbolics and SymPy backends.
- Use `L_show`/`l_show` for LaTeX display.
- Apply formatter helpers for matrix styling.


```julia
using LAlatex
using LaTeXStrings
```

## Backend switching

```julia
LAlatex.@syms x y
set_backend!(:sympy)
LAlatex.@syms a :real => true :positive => true
set_backend!(:symbolics)
LAlatex.@syms b
@show x,y,a,b
```

## Symbolics assumptions (metadata)

```julia
LAlatex.@syms t :real => true
assumptions(t)
```

## SymPy helpers

```julia
@syms_sympy p :real => true :positive => true
q = syms_sympy(:q)
```

## LaTeX helpers

```julia
to_latex(3//4)
```

### to_latex with Symbolics and SymPy

```julia
set_backend!(:symbolics)
LAlatex.@syms s
println( to_latex(s) )

set_backend!(:sympy)
sp = syms(:s; backend=:sympy, real=true)
println(to_latex(sp))
```

## Mixed matrices

Use `mixed_matrix` or `@mixed_matrix` when combining Symbolics/SymPy symbols with complex rationals.
The macro keeps the `[ ... ]` syntax but avoids Julia promotion errors.


```julia
# Example: Symbolics mixed matrix
set_backend!(:symbolics)
x, y = syms(:x, :y)
F = @mixed_matrix [1//2 x; (1 + im)//3 2*y]
l_show(F)
```
