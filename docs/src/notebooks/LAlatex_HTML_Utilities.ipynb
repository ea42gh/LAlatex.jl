## show_html and pr

```julia
show_html("A title"; color="darkred")
pr(L"\qquad " * L"formula: " * L"\sqrt{1+x^2}"; color="olive")
show_html("Title", "with Subtitle"; color="blue", sz1=14, sz2=10, justify="center")
```

## Side-by-side output

```julia
function func1(x)
    println("Output of func1 with argument $x")
end

function func2(y)
    println("Output of func2 with argument $y")
    println("Another line from func2")
end

outputs = [
    capture_output(func1, 1),
    capture_output(func2, 2)
]
show_side_by_side(outputs, ["Function 1", "Function 2"])
```
