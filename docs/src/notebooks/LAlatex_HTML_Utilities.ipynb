# HTML helpers

This notebook demonstrates HTML rendering helpers in LAlatex.

The notebook covers how to

- Produce styled HTML blocks for notebooks.
- Capture and show output side by side.
- Combine HTML helpers with LaTeX output.


```julia
using LAlatex, Latexify, LaTeXStrings
```

## show_html and pr

```julia
display(show_html("A title"; color="darkred"))
display(pr(L"\qquad " * L"formula: " * L"\sqrt{1+x^2}"; color="olive"))
show_html("Title", "with Subtitle"; color="darkgreen", sz1=14, sz2=10, justify="center")
```

## Side-by-side output

```julia
function func1(x)
    println("Output of func1 with argument $x")
end

function func2(y)
    println("Output of func2 with argument $y")
    println("Another line from func2")
end

outputs = [
    capture_output(func1, 1),
    capture_output(func2, 2)
]
show_side_by_side(outputs, ["Function 1", "Function 2"])
```
